```python
score = int(input("점수를 입력하세요: "))

if score >= 90:
    grade = "A" #or print("학점: A")
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print("학점:", grade) 
```
