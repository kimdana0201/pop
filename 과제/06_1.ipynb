```python

import random

target = random.randint(1, 100) #target에 1부터 100까지 숫자가 들어감
attempts = 0

print("1부터 100사이의 숫자를 맞춰보세요!")

while True:
    guess = int(input(f"시도 {attempts}: ")) 
    attempts += 1
    
    if guess < target:
        print("Up!")
    elif guess > target:
        print("Down!")
    else:
        print(f"정답입니다! {attempts}번 만에 맞추셨습니다!")
        break
```
