````python
money = 10000
while True:
    print("\n=== ATM 시스템 ===")
    print("1. 잔액 조회")
    print("2. 입금")
    print("3. 출금")
    print("4. 종료")

    point = input("선택: ")
    
    if point == "1":
        print(f"현재 잔액: {money:,}원")
    elif point == "2":
        a = int(input("입금액: "))
        money += a
        print(f"입금 완료! 현재 잔액: {money:,}원")
    elif point == "3":
        b = int(input("출금액: "))
        if b > money:
            print(f"잔액이 부족합니다. 현재 잔액: {money:,}원")
        else:
            money -= b
            print(f"출금 완료! 현재 잔액: {money:,}원")
    elif point == "4":
        print("이용해 주셔서 감사합니다.")
        break
    else:
        print("잘못된 선택입니다. 1~4번을 선택해주세요.")
````

````python
import random

target = random.randint(1, 100)
attempts = 0

print("1부터 100사이의 숫자를 맞춰보세요!\n")

while True:
    attempts += 1
    guess = int(input(f"시도 {attempts}: "))

    if guess < target:
        print("Up!\n")
    elif guess > target:
        print("Down!\n")
    else:
        print(f"정답입니다! {attempts}번 만에 맞추셨습니다!")
        break
````
