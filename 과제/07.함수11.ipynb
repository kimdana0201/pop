```python
def calculate(a, b, operator):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0:
            return print("0으로 나눌 수 없습니다.")
        else:
            return "잘못된 연산자입니다."
            
    else:
        return "잘못된 연산자입니다." 
print(calculate(10, 3, "+"))
print(calculate(10, 3, "-"))
print(calculate(10, 3, "*"))

print(calculate(10, 0, "/"))
print(calculate(10, 3, "/"))
```
