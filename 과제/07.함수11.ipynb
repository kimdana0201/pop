```python
# 구구단

def print_gugudan(dan, start = 1, end = 9):
    if start > end:
        return "잘못된 범위입니다"
    else:
        for i in range(start, end + 1): #[1, ... 9]
            print(f"{dan} * {i} = {dan * i}")
print_gugudan(5)
print_gugudan(3, 3, 8)
print_gugudan(7, 9, 2)
```
