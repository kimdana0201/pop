```python
# 함수란 무엇인가?
## 믹서기, 자판기

# 파이썬 함수의 구조

# def 함수이름(매개변수):
#     수행할문장1
#     수행할문장2
#     ...

def add(a, b):
    return a + b

result = add(3, 4)  # print(add(3, 4)) 와 같다 
print(result)
```

```python

# 일반적인 함수
## 입력값 : o , 반환값: o 

# def 함수이름(매개변수):
#     수행할문장1
#     수행할문장2
#     ...
#     return 결과값

# 예제 


def add(a, b):
    result = a + b
    return result

a = add(3, 4)
print(a)
```

```python
# 입력값이 없는 함수
## 입력값 : x , 반환값: o
def say():
    return "Hi"

a = say()
print(a)
```

```python
#반환값이 없는 함수
## 입력값 : o , 반환값: x

def add(a, b):
    print(f"{a}, {b}의 합은 {a+b}이다")

a = add(3, 4)
print(a)
```

```python
# 입력값 : o, 반환값 : x

def say():
    print("Hi")

a = say()
print(a)
```

```python
# 매개변수 지정하여 호출하기
def sub(a, b):
    return a - b

result1 = sub(7, 3)
print(result1)

result2 = sub(b=7, a=3)
print(result2)
```

```python
#입력값이 몇 개가 될지 모를 때느 어떻게 해야 할까?

# # def 함수이름(*args):
#     수행할문장1
#     수행할문장2
#     ...

def add_many(*args):    #args = [1, 2, 3, 4, 5]
    result = 0
    for i in args:
        result = result + i
    return result

a = add_many(1, 2, 3, 4, 5)
print(a)
```
