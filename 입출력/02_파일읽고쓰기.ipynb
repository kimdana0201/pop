```python
# 파일객체 = open(파일이름, 파일열기모드)
# r - 읽기모드: 파일을 읽기만 할 때
# w - 쓰기모드: 파일의 내용을 쓸 때
# a - 추가모드: 파일의 마지막에 내용을 추가 할 때
```

```python
f = open("./data/새파일.txt", "w")    #./ - 현재 폴더 - 도수명령어. 관려 공부 더 하기

for i in range(1, 11):
    data = f"{i}번째 줄입니다.\n"
    f.write(data) #객체에 계속 쓸 것 

f.close()  #닫아주는게 프로그램적으로 좋음 
```

```python

f = open("data/안녕하세요.txt", "r")
while True:  #여러줄을 읽어야 돼서 while문 사용 그리고 무조건 True
    line = f.readline()
    if not line: break
    print(line)

f.close()
```

```python
#readlines() 함수 이용하기

f = open("data/안녕하세요.txt", "r")
lines = f.readlines()

for line in lines:
    print(line, end="\n") #를 지운 이유 + end=""를 안하면 2줄이 띄어짐
f.close()
#print 자체애 줄 바꿈이 있다


# print(type(lines))
# print(len(lines))
```

```python
#read 함수 사용하기

f = open("../안녕하세요.txt", "r") #상위폴더로 가는 법  ..
data = f.read()
print(data)
f.close()
```

```python
#파일에 새로운 내용 추가하기
f = open("../안녕하세요.txt", "a")

for i in range(21, 31):
    f.write(f"{i}번째 줄입니다.\n")

f.close()
```

```python
# with문과 함께 사용하기
with open("../안녕하세요.txt", "r") as f:  #상위폴더 -> 상위폴더는  ../../ 사용,  as = f만들어진 것의 이름을 f 라고 하겠다
    f.write()
```
