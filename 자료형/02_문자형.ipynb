```python
# \를 이용해서 작은따옴표와 큰따옴표를 문자열에 포함
food = "python\'s favorite \n is \tperl"
print(food)

# \n 문자 줄을 바꿀 떄
# \t 탭간격을 줄 때 
# \\ \를 그대로 사용할 때
# \' 작은 따옴표를 그대로 표현할 때
```

```python
# 문자열 인덱싱
a = "life is too short"
print(a[8])
print(a[-1])
print(a[-2])
```

```python
# 문자열 슬라이싱
a = '''life is too short, you need Python'''

print(a[0:4])
print(a[12:17])
print(a[5:])
print(a[:17])
print(a[:])
```
