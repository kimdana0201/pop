```python
# 딕셔너리는 어떻게 만들까?

dic = {"name" : "dana", 
       "phone" : "010-4023-8767", 
       "birth" : "00-02-01"
}
dic
#콜론을 기준으로 앞이 이름 뒤가 값. 두번째 요소를 적기위해 , 사용
```

```python
# 딕셔너리 쌍 추가, 삭제하기

dic["hobby"] = "game"
dic
```

```python
dic[3] = [1, 2, 3]
dic
```

```python
# 딕셔너리 요소 삭제하기

del dic[3]
dic
```

```python
del dic["hobby"]
dic
```

```python
# 딕셔너리 요소 변경하기
dic["name"] = "python"
dic
```

```python
# 딕셔너리에서 Key를 사용해서 Value 얻기

print(dic["name"])
print(dic["phone"])
print(dic["birth"])
print(dic["hobby"])
```

```python
# 딕셔너리 관련 함수
## Key 리스트 만들기 - keys

dic = {"name" : "dana", 
       "phone" : "010-4023-8767", 
       "birth" : "00-02-01"
}

dic.keys()
list(dic.keys())  # -> 리스트 형태의 데이터 type으로 바뀜
```

```python
# Vlaue 리스트 만들기 - Values 
li = dic.values()
list(li) #list 안에 넣으면 값들만 정확히 가지고 올 수 있다.
```

```python
# Key, Value 쌍 얻기 - items
li = dic.items()
list(li)
```

```python
# Key: Value 쌍 모두 지우기 - clear
dic.clear()
dic
```

```python
# Key로 Value 얻기 - get

dic = {"name" : "dana", 
       "phone" : "010-4023-8767", 
       "birth" : "00-02-01"
}

dic["name"]
print(dic.get("name"))
print(dic.get("phone"))
print(dic.get("brith"))
```

```python
# 해당 Key가 딕셔너리 안에 있는지 조사하기 - in 

dic = {"name" : "dana", 
       "phone" : "010-4023-8767", 
       "birth" : "00-02-01"
}

print("name" in dic)
print("hobby" in dic)
```

```python
# Keyfh Value 얻기 - pop

dic = {"name" : "dana", 
       "phone" : "010-4023-8767", 
       "birth" : "00-02-01"
}
phone = dic.pop('phone')
print(phone)
dic
```
