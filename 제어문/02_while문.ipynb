```python
#while문의 기본구조

# while 조건문: 조건문은 True 여야한다 False(0)일 경우 반복 수행이 안된다.
#     수행할 문장1
#     수행할 문장2
#     수행할 문장3
#     수행할 문장4
#     ...
```

```python
tree_hit = 0
while tree_hit < 10: #10보다 작을 경우 계속 반복을 돌린다.
    tree_hit = tree_hit + 1 # tree_hit += 1 우측과 같은 뜻
    print(f"나무를 {tree_hit}번 찍었습니다.")
    
    if tree_hit == 10:
        print("나무가 넘어갑니다.")
```

```python
# while 문 만들기
#큰,작은 다옴표 3개는 줄바꿈 가능
prompt = """  
1. Add
2. Del
3. List
4. Quit

Enter number:
"""

number = 0
while number != 4: #4를 넣으면 False가 되기에 끝남.
    print(prompt)
    number = int(input("숫자를 입력하세요(1~4):"))
```

```python
# while문 강제로 빠져 나가기

coffee = 5
money = 300

while money:    #300은 0이 아니기에 실행됨
    print("돈을 받았으니 커피를 줍니다.")
    coffee = coffee -1   #coffee -= 1
    print(f"남은 커피의 양은 {coffee}개입니다.")

    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
        break
```

```python
coffee = 10

while True:    #무한루프문이 됨 -> break가 필요
    money = int(input("돈을 주세요: "))

    if money == 300:
        print("커피를 줍니다.")
        coffee -= 1  # coffee = coffee - 1
    elif money > 300:
        print(f"거스름돈 {money - 300}원을 줍니다.")
        coffee -= 1
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
        print(f"남은 커피의 개수는 {coffee}개 입니다.")

    if coffee == 0:
        print("커피가 다 떨어졌으니, 판매를 중지합니다.")
        break
```
