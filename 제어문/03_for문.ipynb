```python
# for문의 기본구조

# for 변수 in 리스트(또는 튜플, 문자열):
#     수행할_문장1
#     수행할_문장2
```

```python
#전형적인 for문

test_list = ["one", "two","three"]

for i in test_list:
    print(i)
```

```python
# 다양한 for문의 사용

a = [(1,2), (3,4), (5,6)]

for i in a:    #i를 (a, b) or a, b 로 사용가능 튜플이라
    print(i)
```

```python
# for문의 응용

# 5명의 시험 점수
scores = [90, 25, 67, 45, 80]
id = 0

for score in scores:
    id += 1  #id = id + 1 
    if score >= 60:
        print(f"id: {id}, 합격입니다.")
    else:
        print(f"id: {id}, 불합격입니다.")
```

```python
# for문과 continue문

scores = [90, 25, 67, 45, 80]
id = 0

for score in scores:
    id += 1
    
    if score < 60:
        continue

    print(f"id가 {id}인 학생은 합격입니다.")  #2, 4는 실행이 안됨. continue에서 다시 for문 으로 돌가기 때문

# else:
#     print(f"불합격입니다.")    <- 요기 부분 다시 공부
```

```python
# for문과 함께 자주 사용하는 range 함수

a = range(10)
print(a)

b = range(10)
print(list(b))

c = range(1,10)
print(list(c))
print(len(c))
```

```python
# 1부터 10까지 더하는 반복문

hap = 0
for i in range(1, 10+1):  #list에 [1, 2, 3, 4 ... 10] 있는 것
    hap = hap + i # i 에 더해야 1부터 10까지 더하눈고

    print(hap)  # 과정이 보임
print(hap)   #결과 값만 보임
```

```python
# for와 range()를 이용한 구구단을 만들어 보세요.
# 출력예시는
# 2 * 1 = 2
# 2 * 2 + 4
# 2 * 3 = 6
# ...
# 9 * 9 = 81


for i in range(2, 10):
     for k in range(1, 10):
          print(i*k, end="")
     print('')
```

```python
for a in range(2, 10):  #[2, 3, 4, ... 9]
    for b in range(1, 10):   #[1, 2, 3, 4 ... 9]   . a가 2 일 때 9번, 3일 때 9번 반복...
        print(f"{a} * {b} = {a * b}")
```

```python
# enumerate 함수 활용하기
fruits = ["apple", "banana", "orange"]

for fruit  in fruits:
    print(fruit)

for (i, fruit)  in enumerate(fruits):
    print(f"{i}: {fruit}")
```

```python
# zip 함수로 여러 리스트 함께 순회하기
names = ["홍길동", "유재석", "강호동"]
scores = [85, 92, 80]

for name, score in zip(names, scores):
    print(f"{name}: {score}점")
```
