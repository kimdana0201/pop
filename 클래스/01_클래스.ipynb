```python
result = 0

def add(num):
    global result #result는 global로 쓰겠다. ex) 지역변수 a = 10이 있다면 함수 안에서만 쓰이지만 global을 쓴다면 전체 사용 가능. 그러나 오류 찾기가 어려워서 잘 안쓰임
    result += num
    return result
 
print(add(3))
print(add(4))
print(add(5))
```

```python
result1 = 0

def add1(num):
    global result1 
    result1 += num
    return result1
 
print(add1(3))
print(add1(4))
print(add1(5))
```

```python
result2 = 0

def add2(num):
    global result2 
    result2 += num
    return result2
 
print(add2(5))
print(add2(10))
print(add1(20)) #첫번째 계산기에서 더한 20의 값이 나온다
```

```python
class Calculator:
    def __init__(self):   #__ 문법, 무조건 지켜야 됨. self = 자기자신을 가르키는 객체
        self.result = 0

    def add(self, num):
        self.result = self.result + num   #self.result += num
        return self.result   
#1.설계도 만듦

cal1 = Calculator()   # 생성하기 위해 만듦 변수로 받기 위해
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))

print(cal2.add(5))
print(cal2.add(10))
```

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
            result = self.first + self.second
            return result
        
    def sub(self):
            result = self.first - self.second
            return result
        
    def mul(self):
            result = self.first * self.second
            return result
        
    def div(self):
            result = self.first / self.second
            return result
        
cal1 = FourCal()
cal1.setdata(10, 20)
print(cal1.add())
print(cal1.mul())
print(cal1.sub())
print(cal1.div())

cal2 = FourCal()
cal2.setdata(20, 5)
print(cal2.add())
print(cal2.mul())
print(cal2.sub())
print(cal2.div())
```

```python
class FourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        return self.first + self.second
         
    
    def sub(self):
        return self.first - self.second 
    
    def mul(self):
        return self.first * self.second
    
    def div(self):
        return self.first / self.second
    
cal1 = FourCal(10,20)
print(cal1.add())






```

```python
class MoreFourCal(FourCal):
    def pow(self):
        result = self.first ** self.second
        return result
cal4 = MoreFourCal()
cal4.setdata(2, 4)
cal4.pow()
```

```python
# 매서드 오버라이딩

class SafeFourCal(FourCal):
    def div(self):
        if self.second == 0:
            return 0
        
        else:
            return self.first / self.second
        
cal1 = FourCal()
cal1.setdata(4, 0)
# cal1.div() -실행하면 오류남
cal2 = SafeFourCal
cal2.setdata(4, 0)
cal2.div()
```
